```python
import pandas as pd
```

### 2015

```python
df_2015_1 = pd.read_csv('DATOS/2015-03.csv', low_memory=False)
df_2015_2 = pd.read_csv('DATOS/2015-04.csv', low_memory=False)
```

```python
df_2015_1.head(2)
```

```python
df_2015_2.head(2)
```

```python
df_2015_1['Fecha_Retiro'] = pd.to_datetime(df_2015_1['Fecha_Retiro'])
df_2015_1 = df_2015_1[((df_2015_1['Fecha_Retiro'] >= pd.Timestamp('2015-03-29')) & (df_2015_1['Fecha_Retiro'] <= pd.Timestamp('2015-04-05')))]
```

```python
df_2015_2['Fecha_Retiro'] = pd.to_datetime(df_2015_2['Fecha_Retiro'])
df_2015_2 = df_2015_2[((df_2015_2['Fecha_Retiro'] >= pd.Timestamp('2015-03-29')) & (df_2015_2['Fecha_Retiro'] <= pd.Timestamp('2015-04-05')))]
```

```python
df_2015 = pd.concat([df_2015_1, df_2015_2])
```

```python
df_2015.info()
```

```python
del df_2015_1
del df_2015_2
```
